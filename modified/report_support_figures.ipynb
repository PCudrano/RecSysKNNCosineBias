## Figures in support of the report

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from math import pi, sin, cos
```

#### Special case 1

```python
delta = np.linspace(0,4,200)
f = plt.figure()
for n in [2,5,10,100]:
    cos_sim = (n * np.power(delta,2) + 2*delta) / (n * np.power(delta,2) + 2*delta + 1)
    
    plt.plot(delta, cos_sim, label='n={:.0f}'.format(n))

plt.xlabel('delta')
plt.ylabel('cos_sim')
plt.legend()
```

#### Special case 2

```python
delta = np.linspace(0,1,200)
f = plt.figure()
for n in [2,5,10,100]:
    cos_sim = (n * np.power(delta,2) + (n+1)*delta + 1) / ((1+delta) * np.sqrt(n * np.power(delta,2) + 2*delta + 1))
    
    plt.plot(delta, cos_sim, label='n={:.0f}'.format(n))

plt.ylim([0,10.5])
plt.xlabel('delta')
plt.ylabel('cos_sim scaling factor')
plt.legend()
```

```python
delta = np.linspace(0,1.5,200)
f = plt.figure()
for n in [2,5,10,100]:
    cos_sim = 1/np.sqrt(n) * (n * np.power(delta,2) + (n+1)*delta + 1) / ((1+delta) * np.sqrt(n * np.power(delta,2) + 2*delta + 1))
    
    plt.plot(delta, cos_sim, label='n={:.0f}'.format(n))

plt.xlabel('delta')
plt.ylabel('cos_sim')
plt.legend()
```
